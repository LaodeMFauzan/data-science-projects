```python
file_path_location = "../input/catatan_lokasi.csv"

data_location = pd.read_csv(file_path_location, index_col="id")

data_location.groupby('lokasi_dominan').lokasi_dominan.count()
```

```python
file_path_profile = "../input/data_profil.csv"

data_profile = pd.read_csv(file_path_profile, index_col="id")

data_profile.divisi.value_counts()

data_profile.groupby(['divisi','jenis_kelamin']).size()
data_profile.groupby(['divisi','umur']).size()
data_profile.loc[data_profile.loc[:,'divisi']== 'Data Science']
```

```python
left = data_location
right = data_profile

result = left.join(right, how = "inner")
result.loc[result.loc[:,'tanggal'] == '14/06/19'].lokasi_dominan.value_counts()
result.loc[result.loc[:,'lokasi_dominan'] == 'Kota Medan']
result.to_csv ('profile_location_table.csv', index = True)
```
