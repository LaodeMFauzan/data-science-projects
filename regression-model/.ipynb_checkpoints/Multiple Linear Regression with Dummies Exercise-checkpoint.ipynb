You are expected to create a multiple linear regression (similar to the one in the lecture), using the new data. 

In this exercise, the dependent variable is 'price', while the independent variables are 'size', 'year', and 'view'.

#### Regarding the 'view' variable:
There are two options: 'Sea view' and 'No sea view'. You are expected to create a dummy variable for view and include it in the regression

Good luck!

## Import the relevant libraries

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
```

## Load the data

```python
raw_data = pd.read_csv('real_estate_price_size_year_view.csv')
```

```python
raw_data
```

## Create a dummy variable for 'view'

```python
data = raw_data.copy()
```

```python
data['view'] = data['view'].map({'Sea view':1,'No sea view':0})
data
```

```python
data.describe()
```

## Create the regression

### Declare the dependent and the independent variables

```python
y = data['price']
x1 = data[['size','year','view']]
```

### Regression

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data['size'],y)

yhat_no = -5398000 + 2230316 * data['size'] + 27189489 * data['year'] + 56730 * 0
yhat_yes = -5398000 + 2230316 * data['size'] + 27189489 * data['year'] + 56730 * 1

fig = plt.plot(data['size'],yhat_no,lw = 2,c='red')
fig = plt.plot(data['size'],yhat_yes,lw = 2,c='green')

plt.xlabel('size',fontsize = 20)
plt.ylabel('price',fontsize = 20)
plt.show()
```

```python
x
new_data = pd.DataFrame({'const':1,'size':[700,950],'year':[2019,2014],'view':[1,0]})
new_data= new_data[['const','size','year','view']]
new_data
```

```python
new_data.rename(index = {0 : 'House A',1 : 'House B'})
```

```python
predictions = results.predict(new_data)

predictions
```

```python
predictionsdf = pd.DataFrame({'Price Predictions' : predictions})
joined = new_data.join(predictionsdf)
joined.rename(index = {0 : 'House A',1 : 'House B'})
```
